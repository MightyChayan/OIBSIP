# Iris Flower Classification

A machine learning project using the Iris dataset.

This notebook explores the supplied Iris dataset, compares classification algorithms, and evaluates the final model.

```python
from pathlib import Path
import sys

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(PROJECT_ROOT))

from src.train import FEATURE_COLUMNS, load_data, train_and_evaluate

sns.set_theme(style='whitegrid')
```

## 1. Load and Inspect the Data

```python
data = load_data(PROJECT_ROOT / 'Iris.csv')
print(f'Rows: {data.shape[0]}, Columns: {data.shape[1]}')
data.head()
```

```python
data.info()
display(data.describe().round(2))
display(data['Species'].value_counts())
print('Missing values:', data.isna().sum().sum())
```

## 2. Exploratory Data Analysis

```python
sns.pairplot(data[FEATURE_COLUMNS + ['Species']], hue='Species', corner=True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data[FEATURE_COLUMNS].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Feature Correlation Heatmap')
plt.show()
```

### EDA Observations

- The dataset is balanced across all three species.
- Setosa is clearly separated by its short petals.
- Petal length and petal width have a strong positive correlation.
- Versicolor and Virginica overlap slightly, making them the more challenging pair.

## 3. Train, Compare, and Evaluate Models

The shared training module performs a stratified split, cross-validates KNN, Decision Tree, and Random Forest, tunes the strongest candidate, and saves all results.

```python
metrics = train_and_evaluate(
    data_path=PROJECT_ROOT / 'Iris.csv',
    output_dir=PROJECT_ROOT / 'outputs',
)
metrics
```

```python
comparison = pd.read_csv(PROJECT_ROOT / 'outputs' / 'model_comparison.csv')
report = pd.read_csv(PROJECT_ROOT / 'outputs' / 'classification_report.csv', index_col=0)
display(comparison.style.format({c: '{:.3f}' for c in comparison.columns if c != 'model'}))
display(report.round(3))
```

## 4. Conclusion

K-Nearest Neighbors gave the best result for this dataset. The model correctly classified 28 out of 30 test flowers. Petal length and petal width were the most useful measurements for separating the three species.
