# Car Price Prediction

This notebook documents the main analysis behind the car price prediction project. The final user interface is available in `app.py`.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from data_utils import load_data, prepare_features

sns.set_theme(style="whitegrid")
data = load_data()
data.head()
```

```python
print(f"Rows after removing duplicates: {len(data)}")
print(f"Missing values: {data.isna().sum().sum()}")
data.describe().round(2)
```

## Exploratory analysis

The selling-price distribution is right-skewed because the dataset contains many affordable cars and a small number of premium cars.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.histplot(data=data, x="Selling_Price", bins=24, ax=axes[0], color="#1f6f78")
axes[0].set_title("Selling price distribution")
sns.scatterplot(data=data, x="Present_Price", y="Selling_Price", ax=axes[1], color="#ef7d3c")
axes[1].set_title("Original price vs resale price")
plt.tight_layout();
```

```python
data.groupby("Fuel_Type")["Selling_Price"].agg(["count", "median", "mean"]).round(2)
```

## Train and evaluate

The training script compares Random Forest, Extra Trees and Gradient Boosting using five-fold cross-validation. It saves the best full preprocessing pipeline to the `models` directory.

```python
from train_model import train_and_save

metrics = train_and_save()
pd.DataFrame(metrics["model_comparison"]).T.sort_values("cv_r2_mean", ascending=False)
```

```python
pd.Series({
    "Selected model": metrics["selected_model"],
    "Test R-squared": metrics["r2"],
    "Mean absolute error (lakh)": metrics["mae"],
    "RMSE (lakh)": metrics["rmse"],
})
```
